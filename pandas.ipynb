```python
marks=[
    [100,50,80,95,45],
    [90,70,50,45,25],
    [100,50,80,95,45],
    [80,30,60,35,42]
    ]
```

```python
subject=['maths','english','science','hindi','computer']
```

```python
marksheet=pd.DataFrame(marks,columns=[subject],index=['shubham','mansi','mohan','rohan'])
```

```python
marksheet
```

```python
marksheet['total']=100
```

```python
marksheet
```

```python
marksheet['total']=marksheet.sum(axis=1)
```

```python
marksheet
```

```python
marksheet.plot(kind='pie',y='total',autopct="%1.1f%%")
```
